```
#Importing data from text file and storing it as DataFrame and then creating columns for all 

from pyspark.sql import functions as F
empDF = spark.read.format("text").load("/FileStore/data/fixedwidth/employee.txt")
empDF1 = (empDF.withColumn("empID",F.trim(F.substring("value",1,8)).cast("int"))
              .withColumn("lastname",F.trim(F.substring("value",9,9)))
              .withColumn("firstName",F.trim(F.substring("value",19,12)))
              .withColumn("emailID",F.trim(F.substring("value",33,30)))
              .withColumn("jobTitle",F.trim(F.substring("value",65,20)))
              .drop("value")
                )
empDF1.printSchema()
empDF1.display(10)
```

```
from pyspark.sql import functions as F
orderSchemaDDL = "orderNumber int, orderDate Date, requiredDate Date, shippedDate Date, status string, comments string, customerNumber int"

orderdetailDF = spark.read.option("header","true").option("inferSchema","true").csv("dbfs:/FileStore/data/cmdata/orderdetail.csv")
oderDF = spark.read.option("header","true").schema(orderSchemaDDL).csv("dbfs:/FileStore/data/cmdata/order.csv")

orderSummaryJoinDF =( orderDF
                         .join(orderdetailDF,"orderNumber","inner")
                         .where(F.col("quantityOrdered") > 0)
                         .select("orderNumber","orderDate","productCode","quantityOrdered","priceEach")
                    )
orderSummaryJoinDF1 = (orderSummaryJoinDF
                      .withColumn("amount",F.round(orderSummaryJoinDF.quantityOrdered * orderSummaryJoinDF.priceEach,0))
                      .withColumnRenamed("orderDate","Odate")
                      )

productSummaryDF = ( orderSummaryJoinDF1
                        .groupBy("productCode","Odate")
                        .agg(F.sum("quantityOrdered").alias("salesQty"), F.sum("amount").alias("salesAmt"))
                )

productSummaryDF.show(10)
                        
```
