```
#Removed the empty Space from display output with the help of filter function

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 = ebookDF.filter(F.col("value") != "")
ebookDF1.printSchema()
ebookDF1.display()
```

```
#Applying transformation to make the sentence in lower format.

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 = ebookDF.withColumn("value",F.lower("value"))
ebookDF1.printSchema()
ebookDF1.display()
```

```
#Splitting the content with the help of explode and split functions.

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 = ebookDF.withColumn("word", F.explode(F.split("value"," ")))
ebookDF1.printSchema()
ebookDF1.display()
```

```
#Dropping the value column from above output

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 = ebookDF.withColumn("word", F.explode(F.split("value"," "))).drop("value")
ebookDF1.printSchema()
ebookDF1.display()
```

```
#Counting the words with the help of groupBy and agg functions.

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 = ebookDF.withColumn("word", F.explode(F.split("value"," "))).drop("value").groupBy("word").agg(F.count("word").alias("Word_Count"))
ebookDF1.printSchema()
ebookDF1.display()
```

```
#Counting the words with the help of groupBy and agg functions and sorting based on descending and limiting top 10 records.

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 =( ebookDF
           .withColumn("word", F.explode(F.split("value"," ")))
           .drop("value")
           .filter(F.col("word") != "")
           .groupBy("word")
           .agg(F.count("word").alias("Word_Count"))
           .orderBy(F.desc("Word_Count"))
           .limit(10)
          )
ebookDF1.printSchema()
ebookDF1.display()
```

```
#Counting the words with the help of groupBy and agg functions and sorting based on descending and limiting top 10 records.
#Saving the output file in CSV format

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 =( ebookDF
           .withColumn("word", F.explode(F.split("value"," ")))
           .drop("value")
           .filter(F.col("word") != "")
           .groupBy("word")
           .agg(F.count("word").alias("Word_Count"))
           .orderBy(F.desc("Word_Count"))
           .limit(10)
          )
ebookDF1.printSchema()
ebookDF1.display()

ebookDF1.write.format("csv").mode("overwrite").save("/output/trg2609/ebookOut")
```

```
#Counting the words with the help of groupBy and agg functions and sorting based on descending and limiting top 10 records.
#Saving the output file in JSON format in five diffrent files using repartition function

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 =( ebookDF
           .withColumn("word", F.explode(F.split("value"," ")))
           .drop("value")
           .filter(F.col("word") != "")
           .groupBy("word")
           .agg(F.count("word").alias("Word_Count"))
           .orderBy(F.desc("Word_Count"))
           .repartition(5)
           #.limit(10)
          )
ebookDF1.printSchema()
ebookDF1.display()

ebookDF1.write.format("json").mode("overwrite").save("/output/trg2609/ebookOut1")
```

```
#Counting the words with the help of groupBy and agg functions and sorting based on descending and limiting top 10 records.
#Saving the output file in JSON format in five diffrent files using coalesce function

from pyspark.sql import functions as F
ebookDF = spark.read.format("text").load("/FileStore/data/ebook/PrideAndPrejudice.txt")
ebookDF1 =( ebookDF
           .withColumn("word", F.explode(F.split("value"," ")))
           .drop("value")
           .filter(F.col("word") != "")
           .groupBy("word")
           .agg(F.count("word").alias("Word_Count"))
           .orderBy(F.desc("Word_Count"))
           .coalesce(5)
           #.limit(10)
          )
ebookDF1.printSchema()
ebookDF1.display()

ebookDF1.write.format("json").mode("overwrite").save("/output/trg2609/ebookOut1")
```
